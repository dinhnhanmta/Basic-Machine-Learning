```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

```python
img = mpimg.imread("girl3.jpg")
plt.imshow(img)
```

```python
#đổi đầu vào thành dạng list của 1 vector 3 chiêu
x = img.reshape(img.shape[0]*img.shape[1],3)

K = 20
kmeans = KMeans(n_clusters=K).fit(x)
```

```python
y_pred = kmeans.predict(x)
x_recovery = np.zeros_like(x)
for i in range(K):
  #thay thế các điểm ảnh ban đầu thành điểm ảnh của center mà chúng thuộc về
  x_recovery[y_pred == i] = kmeans.cluster_centers_[i]
```

```python
x_recovery = x_recovery.reshape((img.shape[0],img.shape[1],3))
plt.imshow(x_recovery)
```

```python
plt.imshow(x_recovery)
imgplot = plt.imshow(x_recovery)
plt.axis('off')
plt.show()
```
